### import the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

### load the data

```python
experiences = [1.5, 2.0, 2.5, 4.0, 5.0, 6.5]
salaries = [15, 20, 30, 45, 70, 75]
```

### exploratory data analysis (EDA)

```python
# set the style of chart to seaborn
plt.style.use('seaborn-v0_8')

# plt.scatter(x_axis_data, y_axis_data, ...)
plt.scatter(experiences, salaries, label="Salary", color="red")

# add the chart information to make it more readable
plt.title("Experience VS Salaries")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

# show the legend
plt.legend()

# show the grid
plt.grid(axis="y", alpha=0.4)
```
