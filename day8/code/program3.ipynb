# Data Visualization
## Line plot

### import required packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

### load the data

```python
# df => Data Frame (two dimensional collection given by pandas)
df = pd.read_csv("Salary_Data.csv")
df.head()
```

```python
# plt.scatter(x_axis_data, y_axis_data, ...)
plt.plot(df["YearsExperience"], df['Salary'], label="Salary", color="green")
plt.plot(df["YearsExperience"], df['Expences'], label="Expences", color="red")

plt.scatter(df["YearsExperience"], df['Salary'], label="Salary", color="green")
plt.scatter(df["YearsExperience"], df['Expences'], label="Expences", color="red")


# add the chart information to make it more readable
plt.title("Experience VS Salaries")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

# show the legend
plt.legend()

# show the grid
plt.grid(axis="y", alpha=0.4)
```
