# Data Visualization
## Bar chart

### import required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

### load the data

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange']
```

### EDA

```python
plt.bar(fruits, counts, label=fruits, color=bar_colors)

plt.title("Fruit Supply Information")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.grid(alpha=0.2)
plt.legend()
```
