# Data Visualization
## Pie Chart

### import required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

### load the data

```python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = [0, 0, 0.1, 0]
```

### EDA

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%', explode=explode, shadow=True)
plt.legend()
```
